```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
import pandas as pd
from scipy import stats
```

```python
#opening the data files
hfhiggs = h5py.File('higgs_100000_pt_1000_1200.h5', 'r')
hfqcd = h5py.File('qcd_100000_pt_1000_1200.h5', 'r')
```

```python
#loading the data from the files into an array so that it can be used in the notebook
higgs_data = np.array(hfhiggs.get('higgs_100000_pt_1000_1200'))
qcd_data = np.array(hfqcd.get('qcd_100000_pt_1000_1200'))
```

```python
#creating a table of the Higgs data
higgs_table = pd.DataFrame(higgs_data)
higgs_table.columns = ['pt', 'eta', 'phi', 'mass', 'ee2', 'ee3', 'd2', 'angularity', 't1', 't2', 't3', 't21', 't32', 'KtDeltaR']
higgs_table
```

```python
#creating a table of the background data
qcd_table = pd.DataFrame(qcd_data)
qcd_table.columns = ['pt', 'eta', 'phi', 'mass', 'ee2', 'ee3', 'd2', 'angularity', 't1', 't2', 't3', 't21', 't32', 'KtDeltaR']
qcd_table
```

```python
#getting the mean and standard deviation of the two first columns of both the Higgs and QCD data
higgs_mean = np.mean(higgs_table['pt'])
print('Higgs mean: ' + str(higgs_mean))
higgs_std = np.std(higgs_table['pt'])
print('Higgs std: ' + str(higgs_std))
qcd_mean = np.mean(qcd_table['pt'])
print('QCD mean: ' + str(qcd_mean))
qcd_std = np.std(qcd_table['pt'])
print('QCD std: ' + str(qcd_std))
```

```python
print(np.max(qcd_table['pt']))
```

```python
#plotting the p_T histogram of Higgs and comparing it to that of QCD
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
#looping through the different subplots and the momenta from the datasets
for (ax, p, title) in zip(axes, [higgs_table['pt'], qcd_table['pt']], ['Higgs', 'QCD']):
    ax[0].set_title('Transverse momentum histogram')
    ax[0].hist(p, bins=75, density=True, label=title)
    ax[0].set_xlabel('Transverse momentum')
    ax[0].set_ylabel('Number of counts (normalized)')
    ax[0].legend()
    ax[1].set_title('Transverse momentum histogram on a log scale')
    ax[1].hist(p, bins=75, density=True, label=title)
    ax[1].set_xlabel('Transverse momentum')
    ax[1].set_ylabel('Number of counts (log)')
    ax[1].set_yscale('log')
    ax[1].legend()
    
#plt.savefig()
```

```python
#plotting the eta histogram of Higgs and comparing it to that of QCD
fig, axes = plt.subplots(2, 2, figsize=(14, 12))
#looping through the different subplots and the etas from the datasets
for (ax, eta, title) in zip(axes, [higgs_table['eta'], qcd_table['eta']], ['Higgs', 'QCD']):
    ax[0].set_title('Pseudorapidity histogram')
    ax[0].hist(eta, bins=75, density=True, label=title)
    ax[0].set_xlabel('Pseudorapidity')
    ax[0].set_ylabel('Number of counts (normalized)')
    ax[0].legend()
    ax[1].set_title('Pseudorapidity histogram on a log scale')
    ax[1].hist(eta, bins=75, density=True, label=title)
    ax[1].set_xlabel('Pseudorapidity')
    ax[1].set_ylabel('Number of counts (normalized)')
    ax[1].set_yscale('log')
    ax[1].legend()
    
plt.savefig('eta_hist.png')
```

```python
#plotting the eta vs momentum of Higgs and comparing it to that of QCD
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
#looping through the different subplots and the datasets
for (ax, data, title) in zip(axes, [higgs_table, qcd_table], ['Higgs', 'QCD']):
    ax.set_title('Eta as a function of transverse momentum')
    ax.scatter(data['pt'], data['eta'], label=title, s=2)
    ax.set_xlabel('Transverse momentum')
    ax.set_ylabel('Pseudorapidity')
    ax.legend()
    
plt.savefig('eta_v_pt.png')
```

```python
#creating histograms of the rest of the columns in the data
fig, axes = plt.subplots(3, 2, figsize=(14, 16))
#looping through each subplot and the remaining columns to each of them
for ax, column in zip(axes, higgs_table.columns[2:5]):
    ax[0].set_title('{} histogram of Higgs and QCD data'.format(column))
    ax[0].hist(higgs_table[column], bins=75, label='Higgs', weights=np.ones(len(higgs_table[column])) / len(higgs_table[column]))
    ax[0].hist(qcd_table[column], bins=75, label='QCD', weights=np.ones(len(qcd_table[column])) / len(qcd_table[column]), alpha=0.5)
    ax[0].set_xlabel(column)
    ax[0].set_ylabel('Proportion of counts')
    ax[0].legend()
    ax[1].set_title('{} histogram of Higgs and QCD data on a log scale'.format(column))
    ax[1].hist(higgs_table[column], bins=75, label='Higgs', weights=np.ones(len(higgs_table[column])) / len(higgs_table[column]))
    ax[1].hist(qcd_table[column], bins=75, label='QCD', weights=np.ones(len(qcd_table[column])) / len(qcd_table[column]), alpha=0.5)
    ax[1].set_yscale('log')
    ax[1].set_xlabel(column)
    ax[1].set_ylabel('Proportion of counts (log)')
    ax[1].legend()
plt.savefig('higgs_qcd_hist1.png'.format(column))
```
